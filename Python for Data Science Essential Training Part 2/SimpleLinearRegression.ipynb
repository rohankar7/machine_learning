```python
# Linear Regression

# Assumptions
#1 All variables are continuous numeric, not categorical
#2 Data is free from missing values and outliers
#3 There's a linear relationship between predictors and predictant
#4 All predictors are independent of each other
#5 Residuals (or prediction errors) are normally distributed
```

```python
# Simple Linear Regression

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn

from pylab import rcParams

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import scale
```

```python
%matplotlib inline

rcParams['figure.figsize'] = 10,8
```

```python
rooms = 2*np.random.rand(100, 1)+3

rooms[0:10]
```

```python
price = 265 + 6*rooms+abs(np.random.rand(100, 1))
price[:10]
```

```python
plt.plot(rooms,price, 'r^')
plt.xlabel("# of Rooms, 2019 Average")
plt.ylabel("2019 Average Home, 1K USD")

plt.show()
```

```python
X = rooms
y = price

lm = LinearRegression()
lm.fit(X,y)

print(lm.intercept_, lm.coef_)
```

```python
# R squared value for how well the model performs

print(lm.score(X,y))
```
